```python
#11. Write a python program to find the factorial of a number.
```

```python
value = int(input(" Please enter any Value : "))
fact = 1
i = 1

while(i <= value):
    fact = fact * i
    i = i + 1

print("The Result of %d  = %d" %(value, fact))
```

```python
#12. Write a python program to find whether a number is prime or composite.
```

```python

num = int(input("Enter any number : "))
if num > 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(num, "is NOT a prime number")
            break
    else:
        print(num, "is a PRIME number")
elif num == 0 or 1:
    print(num, "is a neither prime NOR composite number")
else:
    print(num, "is NOT a prime number it is a COMPOSITE number")
```

```python
#13. Write a python program to check whether a given string is palindrome or not.
```

```python
string = input("Please enter your own Text : ")
str1 = ""

for i in string:
    str1 = i + str1  
print("Reverse Order :  ", str1)

if(string == str1):
   print("This is a Palindrome String")
else:
   print("This is Not the palindrome string")
```

```python
#14. Write a Python program to get the third side of right-angled triangle from two given sides.
```

```python
def pythagoras(stret,base,cross):
        if stret == str("x"):
            return ("stret = " + str(((cross**2) - (base**2))**0.5))
        elif base == str("x"):
            return ("base = " + str(((cross**2) - (stret**2))**0.5))
        elif cross == str("x"):
            return ("cross= " + str(((stret**2) + (base**2))**0.5))
        else:
            print('solve the problem')
    
print(pythagoras(3,4,'x'))
print(pythagoras(3,'x',5))
print(pythagoras('x',4,5))
print(pythagoras(3,4,5))
```

```python
#15. Write a python program to print the frequency of each of the characters present in a given string.
```

```python
input_string = "Data Science"

frequency_table = {} 
  
for char in input_string: 
    frequency_table[char] = frequency_table.get(char, 0) + 1

    print ("Character frequency table for '{}' is :\n {}".format(input_string, str(frequency_table)))
```
